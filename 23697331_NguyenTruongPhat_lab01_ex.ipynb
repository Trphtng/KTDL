## NUMPY

Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

```python
import numpy as np
from numpy import random
a = random.randint(0, 1000, size=(100))
print(a)
print("10 phần tử đầu tiên" ,a[0:10])
print("5 phần tử cuối cùng",a[-5:])
print("Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100",a[(a % 5 == 0) & (a > 100)])
```

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

```python
len(a[a<100])
```

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

```python
print("Tổng mảng",a.sum())
print("Giá trị trung bình",a.mean())
print("Giá trị trung vị",np.median(a))
print('Tứ phân vị [Q1, Q2, Q3]: ', np.quantile(a, [0.25, 0.5, 0.75]))
print('Phân vị 20 (median): ', np.quantile(a,0.2))
```

### 10. Tính toán các giá trị sau của  thuộc tính giá ('Price')

```python
print('Giá trung bình các xe',car['Price'].mean())
print('Giá trị trung vị các xe',car['Price'].median())
print('Miền giá trị của xe',car['Price'].max()-car['Price'].min())
print('Miền phân vị của xe',car['Price'].quantile(0.75) -car['Price'].quantile(0.25))
print('Điểm phân vị 0.4 và 0.6 của giá',car['Price'].quantile([0.4,0.6]))
```
